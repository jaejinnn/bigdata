### Q2 dplyr 함수들을 사용하여, 다이아몬드의 색 별로 평균 분산을 구해보세요. (Hint: group_by, summarize 사용)

### Q3 가격이 5000 이하인 다이아몬드들에 대해서, 다이아몬드 색에 따른 가격의 평균과 분산 정보를 Bar plot으로 그려보세요 (bar plot 의 x축은 색, y축은 색에따른 평균가격). 해당 bar plot에 표준편차를 errorbar 로 나타내세요. (Hint: ggplot에서 geom_errorbar 사용)



```r
diamonds %>% group_by(color) %>%
  filter(price <= 5000) %>%
  summarize(price_mean = mean(price, na.rm = TRUE), price_sd = sd(price, na.rm = TRUE))

diamonds %>% group_by(color) %>%
  filter(price <= 5000) %>%
  summarize(price_mean = mean(price, na.rm = TRUE), price_sd = sd(price, na.rm = TRUE)) %>%
  ggplot(mapping = aes(x= color, y = price_mean)) +
    geom_bar(stat = "identity", position = "dodge") +
    geom_errorbar(mapping = aes(x = color, ymin = price_mean - price_sd, ymax = price_mean + price_sd), width = 0.4, colour = "skyblue", alpha = 0.9, size = 1.3)
```

### Q4 theme 을 사용해서 그래프를 잘 그려보세요. 다음 조건은 만족해야 합니다.
  - 그래프 안에 글씨 크기들을 기본형보다 크게 그려주세요.
  - 어떤 theme 이든 사용해주세요.

```r
# Q3의 그래프에 theme을 사용

diamonds %>% group_by(color) %>%
  filter(price <= 5000) %>%
  summarize(price_mean = mean(price, na.rm = TRUE), price_sd = sd(price, na.rm = TRUE)) %>%
  ggplot(mapping = aes(x= color, y = price_mean)) +
    geom_bar(stat = "identity", position = "dodge") +
    geom_errorbar(mapping = aes(x = color, ymin = price_mean - price_sd, ymax = price_mean + price_sd), width = 0.4, colour = "skyblue", alpha = 0.9, size = 1.3) +
      theme_bw() +
      theme(axis.line = element_line(colour = "gray"), panel.border = element_blank(), text = element_text(size = 30), axis.text.x = element_text(angle = 45, vjust = 0.5))
```

```r
# 새로운 그래프를 그려서 theme을 사용

ggplot(data = diamonds) +
  geom_bar(mapping = aes(x = color, fill = clarity)) + theme_bw() +
    theme(axis.line = element_line(colour = "black"), text = element_text(size = 20), axis.text.x = element_text(angle = 30, vjust = 0.5))
```
